## pixel만 찍어보기 

csv에서 어디서부터 다음 열로 넘어가는지 모름. 이걸 여러번 테스트 중.

```python
#새로운 테스트

x_pix = [0 for i in range(290421)]
y_pix = [0 for i in range(290421)]

for k in range(0,449):
    for i in range(0, 644):
        j = k*644+i
        x_pix[j] = i
        #print(x_pix[j])
        y_pix[j] = k
        #print(y_pix[j])
        #print(j)
#print('x',x_pix)
#print('y',y_pix)

data = pd.read_csv('testpts.csv')

r = data['r']
g = data['g']
b = data['b']

#290421 빈 배열 만들어주기
fin_ccc = [0 for i in range(290421)]

cc_hex = [hex(cc[i])[2:] for i in range(290421)]
for i in range(290421):
    l=len(cc_hex[i])
    o=''
    for k in range(0, 6-l):   #6자리 안되면 0 채우기
        o = o + '0'
    fin_cc = '#'+ o + cc_hex[i]
    fin_ccc[i] = fin_cc
print(fin_ccc)


##그리기
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)

ax.scatter(x_pix,y_pix, marker='o', s=10, c=fin_ccc)

ax.set_xlabel('x')
ax.set_ylabel('y')
plt.xlim([0,800])
plt.ylim([0,800])

plt.show()
```
